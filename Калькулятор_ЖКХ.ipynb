```python
# ФИКСИРОВАННЫЕ ДАННЫЕ

#Газ
gas = 400.21
# Платежи управляющей компании
uk = 587.93
#Капитальный ремонт
overhaul = 423.58
#Отопление
heat = 1421.35
# Вывоз мусора 
rubbish = 223.9

# Сумма
total_static = gas + uk + overhaul + heat + rubbish

print('Газ: ', gas, 'руб.')
print('Платежи управляющей компании: ', uk, 'руб.')
print('Капитальный ремонт: ', overhaul, 'руб.')
print('Отопление: ', heat, 'руб.')
print('Вывоз мусора: ', rubbish, 'руб.')

# ИЗМЕНЯЕМЫЕ ДАННЫЕ

hot_water_index = 141.12
cold_water_index = 15.8
sewerage_index = 11.96
electricity_index = 3.96

# Горячая вода
hot_water_value = int(input('Введите текущие показатели по счетчику горячей воды: ')) - int(input('Введите предыдущие показатели по счетчику горячей воды: '))
hot_water_total = hot_water_value * hot_water_index
print('Горячая вода: ', round(hot_water_total, 2), 'руб.', '\n')

# Холодная вода
cold_water_value = int(input('Введите текущие показатели по счетчику холодной воды: ')) - int(input('Введите предыдущие показатели по счетчику холодной воды: '))
cold_water_total = cold_water_value * cold_water_index
print('Холодная вода: ', round(cold_water_total, 2), 'руб.', '\n')

# Стоки

sewerage_total = (cold_water_value + hot_water_value) * sewerage_index
print('Стоки: ', round(sewerage_total, 2), 'руб.', '\n')

#Электроэнегргия
electricity_value = int(input('Введите текущие показатели по счетчику электроэнергии: ')) - int(input('Введите предыдущие показатели по счетчику электроэнергии: '))
electricity_total = electricity_value * electricity_index
print('Электроэнергия: ', round(electricity_total, 2), 'руб.', '\n', '\n')

# Итого
total = total_static + hot_water_total + cold_water_total + sewerage_total + electricity_total
print('Итого: ', round(total, 2), 'руб.')
```
